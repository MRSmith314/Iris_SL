# Build Supervised Learning training model

```python
 # Get the target variables. 
target = iris["species"]
target_names = ["0", "1", "2"]
```

```python
 # Get the features. 
data = iris.drop("species", axis=1)
feature_names = data.columns
data.head()
```

```python
 # Split data into training and testing
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
x = iris.drop('species', axis=1)
y= iris.species
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=6)
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
 # Calculate the classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```

```python
# scikit learn confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predictions)
cm
```

```python
# Confusion Matrix with labels
group_names = ['setosa','versicolor','virginica',
               'setosa','versicolor','virginica',
               'setosa','versicolor','virginica']
group_names2 = ['as setosa','as setosa','as setosa',
               'as versicolor', 'as versicolor','as versicolor',
               'as virginica','as virginica','as virginica']

group_counts = ["{0:0.0f}".format(value) for value in
                cm.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cm.flatten()/np.sum(cm)]
labels = [f"{v1}\n{v2}\n{v3}\n{v4}" for v1, v2, v3, v4 in
          zip(group_names,group_names2,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(3,3)
sns.heatmap(cm, annot=labels, fmt='', cmap='Blues')
```
